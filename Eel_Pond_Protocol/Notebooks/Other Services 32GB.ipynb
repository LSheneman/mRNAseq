```python
import numpy
import glob

datadirs = ['../Data/Other_source_32gb_R2/Part1/',
    '../Data/Other_source_32gb_R2/Part2/',
    '../Data/Other_source_32gb_R2/Part3/',
    '../Data/Other_source_32gb_R2/Part5/']
names = ['Trimming', 'Diginorm', 'Assembly', 'Annotation']
# note: different zones!

def _get_time_all(filename):
    for line in open(filename):
        line = line.rstrip()
        if 'system' in line:
            fields1 = line.split('user')
            user = float(fields1[0])
            fields1b = line.split('system')[0].split()[-1]
            system = float(fields1b)
            walltime = line.split('elapsed')[0].split()[-1]
            walltime = walltime.split(':')
            assert len(walltime)<=3
            if (len(walltime) == 3):
                hours = float(walltime[0])
                minutes = float(walltime[1])
                seconds = float(walltime[2])
            else:
                hours = float(0)
                minutes = float(walltime[0])
                seconds = float(walltime[1])
            wall_seconds = hours*60*60 + minutes*60 + float(seconds)
            return float(system), float(user), float(wall_seconds)
    raise Exception(filename)
    
def get_time_all(filenames):
    aa, bb, cc = 0., 0., 0.
    for filename in filenames:
        a, b, c = _get_time_all(filename)
        aa += a
        bb += b
        cc += c
    return aa, bb, cc


filesPart1 = glob.glob(datadirs[0] + '*')
filesPart2 = glob.glob(datadirs[1] + '*')
filesPart3 = glob.glob(datadirs[2] + '*')
filesPart5= glob.glob(datadirs[3] + '*')

partTimes = {}
partTimes[0] = get_time_all(filesPart1)
partTimes[1] = get_time_all(filesPart2)
partTimes[2] = get_time_all(filesPart3)
partTimes[3] = get_time_all(filesPart5)
```

```python
totalTime = []
for x in range (0 , len(partTimes)):
    temp = partTimes[x]
    
    totalTime.append(temp[2])
print totalTime
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5

ind = np.arange(4)  # the x locations for the groups
width = 0.2       # the width of the bars

fig = plt.figure()
ax = fig.add_subplot(111)
rects1 = ax.bar(ind+1, totalTime, width, color='b', align='center')

# add some
ax.set_ylabel('Time (s)')
ax.set_title('Time to execute each section of The Eel Pond Tutorial')
ax.set_xticks(ind+1)
ax.set_xmargin(1)
ax.set_xticklabels(names)

ax.set_ylim(0, 60000)

def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, '%d'%int(height),
                ha='center', va='bottom')

autolabel(rects1)


#plt.show()
plt.savefig('../Graphs/Other_sources_32gb_R2/partTime.png')
```
