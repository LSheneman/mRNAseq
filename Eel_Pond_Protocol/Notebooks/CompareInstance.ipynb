```python
import numpy
import glob

datadirs1 = ['../Data/AWS_m1.large_R1/Part1/',
    '../Data/AWS_m1.large_R1/Part2/',
    '../Data/AWS_m1.large_R1/Part3/',
    '../Data/AWS_m1.large_R1/Part5/']

datadirs2 = ['../Data/AWS_m2.xlarge/Part1/',
    '../Data/AWS_m2.xlarge/Part2/',
    '../Data/AWS_m2.xlarge/Part3/',
    '../Data/AWS_m2.xlarge/Part5/']

datadirs3 = ['../Data/Other_source_32gb_R2/Part1/',
    '../Data/Other_source_32gb_R2/Part2/',
    '../Data/Other_source_32gb_R2/Part3/',
    '../Data/Other_source_32gb_R2/Part5/']

datadirs4 = ['../Data/Other_source_30gb_R3/Part1/',
    '../Data/Other_source_30gb_R3/Part2/',
    '../Data/Other_source_30gb_R3/Part3/',
    '../Data/Other_source_30gb_R3/Part5/']

names = ['Trimming', 'Diginorm', 'Assembly', 'Annotation']
# note: different zones!

def _get_time_all(filename):
    for line in open(filename):
        line = line.rstrip()
        if 'system' in line:
            fields1 = line.split('user')
            user = float(fields1[0])
            fields1b = line.split('system')[0].split()[-1]
            system = float(fields1b)
            walltime = line.split('elapsed')[0].split()[-1]
            walltime = walltime.split(':')
            assert len(walltime)<=3
            if (len(walltime) == 3):
                hours = float(walltime[0])
                minutes = float(walltime[1])
                seconds = float(walltime[2])
            else:
                hours = float(0)
                minutes = float(walltime[0])
                seconds = float(walltime[1])
            wall_seconds = hours*60*60 + minutes*60 + float(seconds)
            return float(system), float(user), float(wall_seconds)
    raise Exception(filename)
    
def get_time_all(filenames):
    aa, bb, cc = 0., 0., 0.
    for filename in filenames:
        a, b, c = _get_time_all(filename)
        aa += a
        bb += b
        cc += c
    return aa, bb, cc

filesPart1L = glob.glob(datadirs1[0] + '*')
filesPart2L = glob.glob(datadirs1[1] + '*')
filesPart3L = glob.glob(datadirs1[2] + '*')
filesPart5L= glob.glob(datadirs1[3] + '*')

filesPart1xL = glob.glob(datadirs2[0] + '*')
filesPart2xL = glob.glob(datadirs2[1] + '*')
filesPart3xL = glob.glob(datadirs2[2] + '*')
filesPart5xL= glob.glob(datadirs2[3] + '*')

filesPart1RS2 = glob.glob(datadirs3[0] + '*.time')
filesPart2RS2 = glob.glob(datadirs3[1] + '*.time')
filesPart3RS2 = glob.glob(datadirs3[2] + '*.time')
filesPart5RS2 = glob.glob(datadirs3[3] + '*.time')

filesPart1RS3 = glob.glob(datadirs4[0] + '*.time')
filesPart2RS3 = glob.glob(datadirs4[1] + '*.time')
filesPart3RS3 = glob.glob(datadirs4[2] + '*.time')
filesPart5RS3 = glob.glob(datadirs4[3] + '*.time')

partTimes1 = {}
partTimes1[0] = get_time_all(filesPart1L)
partTimes1[1] = get_time_all(filesPart2L)
partTimes1[2] = get_time_all(filesPart3L)
partTimes1[3] = get_time_all(filesPart5L)

partTimes2 = {}
partTimes2[0] = get_time_all(filesPart1xL)
partTimes2[1] = get_time_all(filesPart2xL)
partTimes2[2] = get_time_all(filesPart3xL)
partTimes2[3] = get_time_all(filesPart5xL)

partTimes3 = {}
partTimes3[0] = get_time_all(filesPart1RS2)
partTimes3[1] = get_time_all(filesPart2RS2)
partTimes3[2] = get_time_all(filesPart3RS2)
partTimes3[3] = get_time_all(filesPart5RS2)

partTimes4 = {}
partTimes4[0] = get_time_all(filesPart1RS3)
partTimes4[1] = get_time_all(filesPart2RS3)
partTimes4[2] = get_time_all(filesPart3RS3)
partTimes4[3] = get_time_all(filesPart5RS3)
```
