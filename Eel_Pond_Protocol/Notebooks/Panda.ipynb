```python
#centralized imports
import glob
import numpy as np
import scipy as sp
import matplotlib 
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
import pandas as pd


#function definitions
def get_time(filename):
    for line in open(filename):
        line = line.rstrip()
        if 'system' in line:
            fields1 = line.split('user')
            user = float(fields1[0])
            fields1b = line.split('system')[0].split()[-1]
            system = float(fields1b)
            walltime = line.split('elapsed')[0].split()[-1]
            walltime = walltime.split(':')
            assert len(walltime)<=3
            if (len(walltime) == 3):
                hours = float(walltime[0])
                minutes = float(walltime[1])
                seconds = float(walltime[2])
            else:
                hours = float(0)
                minutes = float(walltime[0])
                seconds = float(walltime[1])
            wall_seconds = hours*60*60 + minutes*60 + float(seconds)
            return float(wall_seconds)

def get_all_categories(cat_series):
    cat_sets = (set(to_cat_list(x)) for x in cat_series) 
    return sorted(set.union(*cat_sets))

def get_english(cat):
    code, names = cat.split('.') 
    if '|' in names:
        names = names.split(' | ')[1] 
    return code, names.strip()

def to_cat_list(catstr):
    stripped = (x.strip() for x in catstr.split(','))
    return [x for x in stripped if x]
```

```python
#Get all file names in the runs of interes
locations = [(glob.glob("../Data/" + field + "/" + part + "/*.vm.txt")) for part in sections  for field in run]
#grab headers
infile = open(locations[0][0])
line = infile.readline()
headers = [(str.split(line))[x] for x in range(len(str.split(line)))]
headers.append('file')
infile.close()
outputFile = open('../tests/TEST.txt', 'wb')
for test in range(len(locations)):
    for part in range(len(locations[test])):
        tempFile = open(locations[test][part])
        for index, line in enumerate(tempFile):
            if 'buff' not in line:
                temp = ''
                line = str.split(line)
                for i in range(len(line)):
                    temp = temp + line[i] + '  '
                temp = temp + locations[test][part] + '\n'
                outputFile.write(temp)
        tempFile.close()
outputFile.close()
```
