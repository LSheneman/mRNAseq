```python
#function definitions
def avg(*args):
    return np.array(args).mean()

def stdDev(*args):
    return np.array(args).std()

def sum(*args):
    return np.array(args).sum()
```

    if 'sy' in field:      #percent of cpu used for system clean up

    if 'us' in field:      #percent of cpu used for programs

    if 'id' in field:      #waiting for new task -- idle

    if 'wa' in field:      #waiting for I/O (bad >25%)
        wa_index = index

    #################   Memory    ############
    if 'si' in field:      #swap into RAM
    	si_index = index
    if 'so' in field:      #swap memory in to VM
        so_index = index
    if 'free' in field:    #free memory
        free_index = index
        
    #################   I/O    ############
    if 'bi' in field:      #data read 1 kb blocks
        bi_index = index
    if 'bo' in field:      #data written 1kb blocks
        bo_index = index

    #################   Disk    ############
    if 'in' in field:      #disk faults
        df_index = index

```python
#Makes structure for data in format {Run : {part): {file}}
import glob
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt

file_names = {}
for field in run:
    tempRun = field
    exec("file_names['" + tempRun + "'] ={}")
    for part in parts:
        subtract = len("../Data/" + field + "/" + part)
        exec("file_names['" + tempRun + "']['" + part + "']= {}")
        temp = glob.glob("../Data/" + field + "/" + part + "/*.vm.txt")
        count = 0
        for index in temp:
            exec("file_names['" + tempRun + "']['" + part + "'][temp[count][subtract:-7]] ={}")
            exec("file_names['" + tempRun + "']['" + part + "'][temp[count][subtract:-7]]['process_waiting_cpu'] ={}")
            exec("file_names['" + tempRun + "']['" + part + "'][temp[count][subtract:-7]]['process_waiting_oth'] ={}")
            exec("file_names['" + tempRun + "']['" + part + "'][temp[count][subtract:-7]]['io_wait_per'] ={}")
            exec("file_names['" + tempRun + "']['" + part + "'][temp[count][subtract:-7]]['swap_out'] ={}")
            exec("file_names['" + tempRun + "']['" + part + "'][temp[count][subtract:-7]]['swap_in'] ={}")
            exec("file_names['" + tempRun + "']['" + part + "'][temp[count][subtract:-7]]['free_mem'] ={}")
            exec("file_names['" + tempRun + "']['" + part + "'][temp[count][subtract:-7]]['disk_faults'] ={}")
            count= count + 1
del temp
```

```python
for field in file_names:
    for key in file_names[field]:
        for filename in file_names[field][key]:
            tempFile = open('../Data/' + field + '/' + key + filename + '.vm.txt')
            
            process_waiting_cpu = []
            process_waiting_oth = []
            io_wait_per = []
            swap_out = []
            swap_in = []
            free_mem = []
            disk_faults = []
            
            for index, line in enumerate(tempFile):
                if 'buff' not in line:
                    temp = str.split(line)
                    process_waiting_cpu.append(float(temp[r_index])) #avg
                    process_waiting_oth.append(float(temp[b_index])) #avg
                    io_wait_per.append(float(temp[wa_index])) #percent
                    swap_out.append(float(temp[so_index])) #per sec
                    swap_in.append(float(temp[si_index])) #per sec
                    free_mem.append(float(temp[free_index])) #amount
                    disk_faults.append(float(temp[df_index])) #avg
                                
            file_names[field][key][filename]['process_waiting_cpu'] = avg(process_waiting_cpu)
            file_names[field][key][filename]['process_waiting_oth'] = avg(process_waiting_oth)
            file_names[field][key][filename]['io_wait_per'] = avg(io_wait_per)
            file_names[field][key][filename]['swap_out'] = avg(swap_out)
            file_names[field][key][filename]['swap_in'] = avg(swap_in)
            file_names[field][key][filename]['free_mem'] = min(free_mem)/pow(10,4)
            file_names[field][key][filename]['disk_faults'] = avg(disk_faults)
                        
```
