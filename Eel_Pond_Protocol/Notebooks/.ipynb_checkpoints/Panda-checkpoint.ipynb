```python
#centralized imports
import glob
import numpy as np
import scipy as sp
import matplotlib 
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
import pandas as pd


#function definitions
def get_time(filename):
    for line in open(filename):
        line = line.rstrip()
        if 'system' in line:
            fields1 = line.split('user')
            user = float(fields1[0])
            fields1b = line.split('system')[0].split()[-1]
            system = float(fields1b)
            walltime = line.split('elapsed')[0].split()[-1]
            walltime = walltime.split(':')
            assert len(walltime)<=3
            if (len(walltime) == 3):
                hours = float(walltime[0])
                minutes = float(walltime[1])
                seconds = float(walltime[2])
            else:
                hours = float(0)
                minutes = float(walltime[0])
                seconds = float(walltime[1])
            wall_seconds = hours*60*60 + minutes*60 + float(seconds)
            return float(wall_seconds)

def get_all_categories(cat_series):
    cat_sets = (set(to_cat_list(x)) for x in cat_series) 
    return sorted(set.union(*cat_sets))

def get_english(cat):
    code, names = cat.split('.') 
    if '|' in names:
        names = names.split(' | ')[1] 
    return code, names.strip()

def to_cat_list(catstr):
    stripped = (x.strip() for x in catstr.split(','))
    return [x for x in stripped if x]
```

```python
from matplotlib.ticker import FuncFormatter
def megabytes(x, pos):
       """Formatter for Y axis, values are in megabytes"""
       return '%1.f' % (x/(1024*1024))
    
def add_bars(parameter, Label, header, *axis):
    (axis[0]).bar(ind, run1[parameter], width, color='r',align='center',label='m2.xl')
    (axis[0]).bar(ind+width, run2[parameter], width, color='g',align='center',label='30GB R2')
    (axis[0]).bar(ind+width*2, run3[parameter], width, color='b',align='center',label='30GB R3')
    #label
    (axis[0]).set_ylabel(Label)
    (axis[0]).set_title(header)
    (axis[0]).set_xticks(ind+width)
    (axis[0]).set_xticklabels(sections, rotation=35, ha = 'right')
    (axis[0]).set_aspect('auto')
  
```
