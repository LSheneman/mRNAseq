```python
#Useful when finding total stats of a run and/or comparing states accross runs 
#Uses *.time files as input
#Variables to pass in run data is assummed to be in ../Data

########### inputs -- files read from disk to RAM, output --files read from RAM to disk  ********PER SECONDS
run = ( 'AWS_m2.xlarge', 'Other_source_32gb_R2','Other_source_30gb_R3')
sections = ('Part1','Part2','Part3','Part5')

part1 = ['adapter_trim', 'trim_run', 'quality_trim', 'extract_pairs', 'trim_clean']
part2= ['diginorm_pe', 'diginorm_se','filter-adund']
part3 = ['trinity_split', 'trinity']
part5 = ['khmer_part','group','format_annotate.sh','nema.x.mouse','mouse.x.nema', 'homology','orthology','get_mouse_info','annotate']
names = ['Trimming', 'Diginorm', 'Assembly', 'Annotation']
```

```python
#centralized imports
import glob
import numpy as np
import scipy as sp
import matplotlib 
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
import pandas as pd


#function definitions
def get_time(filename):
    for line in open(filename):
        line = line.rstrip()
        if 'system' in line:
            fields1 = line.split('user')
            user = float(fields1[0])
            fields1b = line.split('system')[0].split()[-1]
            system = float(fields1b)
            walltime = line.split('elapsed')[0].split()[-1]
            walltime = walltime.split(':')
            assert len(walltime)<=3
            if (len(walltime) == 3):
                hours = float(walltime[0])
                minutes = float(walltime[1])
                seconds = float(walltime[2])
            else:
                hours = float(0)
                minutes = float(walltime[0])
                seconds = float(walltime[1])
            wall_seconds = hours*60*60 + minutes*60 + float(seconds)
            return float(wall_seconds)

def get_all_categories(cat_series):
    cat_sets = (set(to_cat_list(x)) for x in cat_series) 
    return sorted(set.union(*cat_sets))

def get_english(cat):
    code, names = cat.split('.') 
    if '|' in names:
        names = names.split(' | ')[1] 
    return code, names.strip()

def to_cat_list(catstr):
    stripped = (x.strip() for x in catstr.split(','))
    return [x for x in stripped if x]
```

```python
#Get all file names in the runs of interes
locations = [(glob.glob("../Data/" + field + "/" + part + "/*.vm.txt")) for part in sections  for field in run]
#grab headers
infile = open(locations[0][0])
line = infile.readline()
headers = [(str.split(line))[x] for x in range(len(str.split(line)))]
headers.append('file')
infile.close()
outputFile = open('../tests/TEST.txt', 'wb')
for test in range(len(locations)):
    for part in range(len(locations[test])):
        tempFile = open(locations[test][part])
        for index, line in enumerate(tempFile):
            if 'buff' not in line:
                temp = ''
                line = str.split(line)
                for i in range(len(line)):
                    temp = temp + line[i] + '  '
                temp = temp + locations[test][part] + '\n'
                outputFile.write(temp)
        tempFile.close()
outputFile.close()
```

```python
#read in data
result = DataFrame(pd.read_csv('../tests/Test.txt', sep='\s+', names=list(headers)))
```

```python
#combine dataframes
temp = pd.DataFrame(result.file.str.split('/').tolist(),
                                   columns = ['junk','junk','run','part','name'])
temp=temp[['run','part','name']]
master = temp.join(result, how='left')
#master.head()
temp = pd.DataFrame(master.name.str.split('.vm').tolist())
temp = pd.DataFrame(temp[0].str.split('.sh').tolist())
master.name = temp[0]
#master.head()
del temp
```

```python
def add_bars(parameter, Label, header, *axis):
    (axis[0]).bar(ind, run1[parameter], width, color='r',align='center',label='m2.xl')
    (axis[0]).bar(ind+width, run2[parameter], width, color='g',align='center',label='30GB R2')
    (axis[0]).bar(ind+width*2, run3[parameter], width, color='b',align='center',label='30GB R3')
    #label
    (axis[0]).set_ylabel(Label)
    (axis[0]).set_title(header)
    (axis[0]).set_xticks(ind+width)
    (axis[0]).set_xticklabels(sections, rotation=35, ha = 'right')
    (axis[0]).set_aspect('auto')
  
```

```python
#Change for each part
Part = '1'
index = pd.DataFrame(part1, columns=['name'])


working_part = master[master['part'].str.contains(Part)]
test = working_part[['run','name','r','b','free','wa']]
index = index.reset_index()
index = index.set_index('name')
test = test.join(index, on='name')
test = test.sort('index')

grouped = test.groupby(['index','run','name'])
grouped= grouped.agg({'r': 'mean', 'b': 'mean', 'free': min, 'wa':'mean'})
completed = grouped.reset_index()
run1 = completed[completed['run'].str.contains(re1+run[0])]
run2 = completed[completed['run'].str.contains(re1+run[1])]
run3 = completed[completed['run'].str.contains(re1+run[2])]
sections = get_all_categories(run1.name)
completed = grouped.reset_index()
run1 = completed[completed['run'].str.contains(re1+run[0])]
run2 = completed[completed['run'].str.contains(re1+run[1])]
run3 = completed[completed['run'].str.contains(re1+run[2])]
sections = get_all_categories(run1.name)
```

```python
#Graphs
Name = "Trim"
N = len(sections)
ind = np.arange(N)   # the x locations for the groups
width = 0.27     # the width of the bars

# Make a figure with 6 subplots and axes
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(2, 2)
fig.suptitle(Name, fontsize=36)

#subplot 1
parameter = 'r'
header = 'Processes Waiting on CPU'
Label = 'Average number'
add_bars(parameter, Label, header, ax1)
ax1.legend(loc =0)

#subplot 2
parameter = 'wa'
header = 'Time Spent Waiting on I/O'
Label = 'Percent Total Time'
add_bars(parameter, Label, header, ax2)
ax2.legend(loc =2)
#subplot 3
parameter = 'free'
header = 'Minimum Free Memory'
Label = 'Memory (MB)'
add_bars(parameter, Label, header, ax3)
ax3.legend(loc =0)

figure = plt.gcf() # get current figure
figure.set_size_inches(16, 12)
```
