```python
import glob
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt

m1l_dir = ['../Data/AWS_m1.large_R1/Part1/',
    '../Data/AWS_m1.large_R1/Part2/',
    '../Data/AWS_m1.large_R1/Part3/',
    '../Data/AWS_m1.large_R1/Part5/']

m2xl_dir = ['../Data/AWS_m2.xlarge/Part1/',
    '../Data/AWS_m2.xlarge/Part2/',
    '../Data/AWS_m2.xlarge/Part3/',
    '../Data/AWS_m2.xlarge/Part5/']

oth1_dir = ['../Data/Other_source_32gb_R2/Part1/',
    '../Data/Other_source_32gb_R2/Part2/',
    '../Data/Other_source_32gb_R2/Part3/',
    '../Data/Other_source_32gb_R2/Part5/']

oth2_dir = ['../Data/Other_source_30gb_R3/Part1/',
    '../Data/Other_source_30gb_R3/Part2/',
    '../Data/Other_source_30gb_R3/Part3/',
    '../Data/Other_source_30gb_R3/Part5/']

names = ['Trimming', 'Diginorm', 'Assembly', 'Annotation']
# note: different zones!

def _get_max_resident(filename):
    for line in open(filename):
        line = line.rstrip()
        if 'system' in line:
            fields1 = line.split('user')
            user = float(fields1[0])
            fields1b = line.split('system')[0].split()[-1]
            system = float(fields1b)
            walltime = line.split('elapsed')[0].split()[-1]
            walltime = walltime.split(':')
            assert len(walltime)<=3
            if (len(walltime) == 3):
                hours = float(walltime[0])
                minutes = float(walltime[1])
                seconds = float(walltime[2])
            else:
                hours = float(0)
                minutes = float(walltime[0])
                seconds = float(walltime[1])
            wall_seconds = hours*60*60 + minutes*60 + float(seconds)
            return float(system), float(user), float(wall_seconds)
    raise Exception(filename)
```

```python
file_names = {}

file_names['m1l_dir'] = {}
file_names['m2xl_dir']= {}
file_names['oth1_dir'] = {}
file_names['oth2_dir'] = {}

file_names['m1l_dir']['Part1'] = []
file_names['m1l_dir']['Part2'] = []
file_names['m1l_dir']['Part3'] = []
file_names['m1l_dir']['Part5'] = []

file_names['m2xl_dir']['Part1'] = []
file_names['m2xl_dir']['Part2'] = []
file_names['m2xl_dir']['Part3'] = []
file_names['m2xl_dir']['Part5'] = []

file_names['oth1_dir']['Part1'] = []
file_names['oth1_dir']['Part2'] = []
file_names['oth1_dir']['Part3'] = []
file_names['oth1_dir']['Part5'] = []

file_names['oth2_dir']['Part1'] = []
file_names['oth2_dir']['Part2'] = []
file_names['oth2_dir']['Part3'] = []
file_names['oth2_dir']['Part5'] = []

file_names['m1l_dir']['Part1'] = glob.glob(m1l_dir[0] + '*')
file_names['m1l_dir']['Part2'] = glob.glob(m1l_dir[1] + '*')
file_names['m1l_dir']['Part3'] = glob.glob(m1l_dir[2] + '*')
file_names['m1l_dir']['Part5'] = glob.glob(m1l_dir[3] + '*')

file_names['m2xl_dir']['Part1'] = glob.glob(m2xl_dir[0] + '*')
file_names['m2xl_dir']['Part2'] = glob.glob(m2xl_dir[1] + '*')
file_names['m2xl_dir']['Part3'] = glob.glob(m2xl_dir[2] + '*')
file_names['m2xl_dir']['Part5'] = glob.glob(m2xl_dir[3] + '*')

file_names['oth1_dir']['Part1'] = glob.glob(oth1_dir[0] + '*')
file_names['oth1_dir']['Part2'] = glob.glob(oth1_dir[1] + '*')
file_names['oth1_dir']['Part3'] = glob.glob(oth1_dir[2] + '*')
file_names['oth1_dir']['Part5'] = glob.glob(oth1_dir[3] + '*')

file_names['oth2_dir']['Part1'] = glob.glob(oth2_dir[0] + '*.time')
file_names['oth2_dir']['Part2'] = glob.glob(oth2_dir[1] + '*.time')
file_names['oth2_dir']['Part3'] = glob.glob(oth2_dir[2] + '*.time')
file_names['oth2_dir']['Part5'] = glob.glob(oth2_dir[3] + '*.time')
```

```python
file_stats = {}

file_stats['m1l_dir'] = {}
file_stats['m2xl_dir']= {}
file_stats['oth1_dir'] = {}
file_stats['oth2_dir'] = {}

file_stats['m1l_dir']['Part1'] = {}
file_stats['m1l_dir']['Part2'] = {}
file_stats['m1l_dir']['Part3'] = {}
file_stats['m1l_dir']['Part5'] = {}

file_stats['m2xl_dir']['Part1'] = {}
file_stats['m2xl_dir']['Part2'] = {}
file_stats['m2xl_dir']['Part3'] = {}
file_stats['m2xl_dir']['Part5'] = {}

file_stats['oth1_dir']['Part1'] = {}
file_stats['oth1_dir']['Part2'] = {}
file_stats['oth1_dir']['Part3'] = {}
file_stats['oth1_dir']['Part5'] = {}

file_stats['oth2_dir']['Part1'] = {}
file_stats['oth2_dir']['Part2'] = {}
file_stats['oth2_dir']['Part3'] = {}
file_stats['oth2_dir']['Part5'] = {}
```

```python
for field in file_names:
    for key in file_names[field]:
         for filename in file_names[field][key]:
            file_stats[field][key][filename] = {}
            for line in open(filename):
                    line = line.rstrip()
                    if 'system' in line:
                        line = line.split('system')
                        line = line[1].split('elapsed')
                        file_stats[field][key][filename]['time'] = line[0].split(' ')[1]
                        line = line[1].split('%CPU')
                        
                        file_stats[field][key][filename]['cpu_usage'] = line[0].split(' ')[1]
                        line = line[1].split('avgdata')
                        line = line[1].split('maxresident')
                        file_stats[field][key][filename]['memory'] = line[0].split(' ')[1]
                    elif 'outputs' in line:
                        line = line.split('inputs')
                        file_stats[field][key][filename]['inputs']= line[0]
                        line = line[1].split('outputs')
                        file_stats[field][key][filename]['outputs'] = line[0].split('+')[1]
                        
                        line = line[1].split('major')
                        file_stats[field][key][filename]['major'] = line[0].split(' (')[1]
                        
                        line = line[1].split('minor')
                        file_stats[field][key][filename]['minor'] = line[0].split('+')[1]
```

```python
def avg(*args):
    return np.array(args).mean()

def stdDev(*args):
    return np.array(args).std()
```

```python
run = 'm1l_dir'
part ='Part1'
totals = {}
totals['memory'] = []
totals['cpu_usage'] = []
totals['inputs'] = []
totals['outputs'] = []
count = 0
for key in file_stats[run][part]:
    totals['memory'].append(float(file_stats[run][part][key]['memory']))
    totals['cpu_usage'].append(float(file_stats[run][part][key]['cpu_usage'])/4)
    totals['inputs'].append(float(file_stats[run][part][key]['inputs']))
    totals['outputs'].append(float(file_stats[run][part][key]['outputs']))
    count+=1

#m1_large
print 'm1.large', part

print 'Avg input: ', avg(totals['inputs']), '     Standard_Dev: ', stdDev(totals['inputs'])
print 'Avg output: ', avg(totals['outputs']), '     Standard_Dev: ', stdDev(totals['outputs'])
print 'Avg cpu_usage: ', avg(totals['cpu_usage']), '     Standard_Dev: ', stdDev(totals['cpu_usage'])
print 'Avg memory: ', avg(totals['memory'])/pow(10,6), '     Standard_Dev: ', stdDev(totals['memory'])/pow(10,6)
print max(totals['memory'])/pow(10,6)
```
